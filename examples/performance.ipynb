## Preparation of data

```python
df2 = pd.read_csv("../datasets_tests/monomers_original.csv", sep=",")
df2
```

```python
perf = Performance(df1,df2)
```

## Calculation of performance

It performs (number of calculated / number of not calculated)*100


```python
perf.calculate_performance(df1_index=['smiles_polymer','mechanism'],df2_index=['name','smiles','classes','smiles_polymer'])
```
